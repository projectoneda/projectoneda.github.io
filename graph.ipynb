```python
year_stock_data = pd.read_csv('csv/year_stock_data.csv')

year_date = year_stock_data.Date
year_open = year_stock_data.Open
```

```python
# JP Morgan 
jp = pd.read_csv('csv/jp.csv')
jp_date = jp.Date
jp_open = jp.Open

fig, ax1 = plt.subplots(figsize=(15,5), facecolor=('#35424A'))

color = '#E8491D'
ax1.plot(jp_date, jp_open, color=color)
ax1.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=False)
ax1.set_yticklabels([])

ax2 = ax1.twinx() 

color = '#35424A'
ax2.plot(year_date, year_open, color=color)
ax2.set_yticklabels([])
```
